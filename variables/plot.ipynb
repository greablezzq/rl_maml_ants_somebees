```python
import matplotlib.pyplot as plt
import numpy as np
import tqdm
```

```python
ants_SimpleDDQN_training_episode_reward = np.load('ants_SimpleDDQN_training_episode_reward.npy')
ants_DDQN_training_episode_reward_together = np.load('ants_DDQN_training_episode_reward_together.npy')
ants_DDQN_training_episode_reward_step_1=np.load('ants_DDQN_training_episode_reward_step_1.npy')
ants_DDQN_training_episode_reward_step_2=np.load('ants_DDQN_training_episode_reward_step_2.npy')
ants_DDQN_training_episode_reward_together_normal=np.load('ants_DDQN_training_episode_reward_together_normal.npy')
ants_DDQN_training_episode_reward_step_1_normal=np.load('ants_DDQN_training_episode_reward_step_1_normal.npy')
ants_MAML_training_last_episode_reward=np.load('ants_MAML_training_last_episode_reward.npy')
ants_comparision_MAML_finetuning=np.load('ants_comparision_MAML_finetuning.npy')
cartpole_comparision_MAML_finetuning=np.load('cartpole_comparision_MAML_finetuning.npy')
cartpole_training_episode_reward=np.load('cartpole_training_episode_reward.npy')
```

```python
kernel = [0.02 for _ in range(50)]
plt.figure()
plt.plot(ants_SimpleDDQN_training_episode_reward, label='Episode reward', linewidth=1)
plt.plot(np.convolve(ants_SimpleDDQN_training_episode_reward, kernel, mode='same'), label='Average episode reward', linewidth=2)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Training for simple DDQN')
plt.legend()
plt.show()
```

```python
f = open("ex1.txt")
lines = f.readlines()
reward=[]
for i in range(300):
    reward.append(float(lines[i].split()[3]))
a=[0.05 for _ in range(20)]
plt.figure()
plt.plot([i*5 for i in range(200)], np.convolve(reward, a,mode='same')[0:200], label='MAML')
plt.plot([i*0.2 for i in range(5000)], np.convolve(cartpole_training_episode_reward,[0.01 for _ in range(100)],mode='same')[0:5000], label='Single Task')
plt.xlabel('Epochs')
plt.ylabel('Reward')
plt.ylim((0,450))
plt.title('Comparision between MAML and single task training')
plt.legend()
plt.show()
```
